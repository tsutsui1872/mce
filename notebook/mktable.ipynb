**Table S2.** Forcing and response parameters for the CMIP5 models. $C_\mathrm{T}$
denotes the total heat capacity converted into an equivalent ocean depth
considering an ocean area fraction of 0.71. The numbers in parentheses are
estimated by the conventional regression method for $\lambda$ and ECS, and
computed from model output for TCR. The ECS and $\lambda$ values estimated by
the conventional method are basically the same as in the previous studies
(Andrews et al., 2012; Forster et al., 2013; Flato et al., 2013), except for
possible dependence on the process of computing anomaly data from model
output. The size of $\alpha$ is generally smaller than that estimated with
$\beta = 1$ as assumed in the conventional method.
